```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sktime.detection.clasp import ClaSPSegmentation

df = pd.read_csv('../data/iot_telemetry_data.csv',
                 names=['ts','device','co','humidity','light',
                        'lpg','motion','smoke','temp'],
                 skiprows=1)
df['ts'] = pd.to_datetime(df['ts'].astype(float), unit='s')
df = df.sort_values(['device', 'ts'])
df['motion'] = df['motion'].astype(bool)
df['light'] = df['light'].astype(bool)
```

```python
device_id = 'b8:27:eb:bf:9d:51'
device_df = df[df['device'] == device_id].copy()
device_df = device_df.set_index('ts')

# Use temperature as our target series
temp_series = device_df['temp']
print(temp_series.shape)
print(temp_series.head())
```

```python
clasp = ClaSPSegmentation(period_length=50, n_cps=5)
clasp.fit(temp_series)
change_points = clasp.predict(temp_series)
print("Detected change points:")
print(change_points)
```

```python
# Use only first 5000 points for demonstration
temp_series_small = temp_series.iloc[:5000]

clasp = ClaSPSegmentation(period_length=50, n_cps=5)
clasp.fit(temp_series_small)
change_points = clasp.predict(temp_series_small)
print("Detected change points:")
print(change_points)
```

```python
fig, ax = plt.subplots(figsize=(14, 4))
ax.plot(temp_series_small.values, label='Temperature', color='steelblue')

for cp in change_points:
    ax.axvline(x=cp, color='red', linestyle='--', alpha=0.7)

ax.set_title(f'Change Point Detection — Device: {device_id} (sample: 5000 pts)')
ax.set_ylabel('Temperature (°C)')
ax.legend()
plt.tight_layout()
plt.savefig('../data/changepoint_detection.png')
plt.show()
```

```python
# Detect motion events as anomalies
motion_series = device_df['motion'].astype(int).iloc[:5000]

fig, axes = plt.subplots(2, 1, figsize=(14, 6), sharex=True)

axes[0].plot(temp_series_small.values, color='steelblue', label='Temperature')
for cp in change_points:
    axes[0].axvline(x=cp, color='red', linestyle='--', alpha=0.7)
axes[0].set_ylabel('Temperature')
axes[0].legend()

axes[1].fill_between(range(len(motion_series)), 
                      motion_series.values, 
                      alpha=0.6, color='orange', label='Motion Event')
axes[1].set_ylabel('Motion Detected')
axes[1].legend()

plt.suptitle('Temperature Change Points vs Motion Events')
plt.tight_layout()
plt.savefig('../data/motion_vs_changepoints.png')
plt.show()
```
